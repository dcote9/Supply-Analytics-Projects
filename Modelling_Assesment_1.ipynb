```python
#!pip install pulp
import pandas as pd
import os
#print ("Current directory is:", os.getcwd())
#os.chdir ("C:/Users/Usuario/Downloads")
#print("Current director is now:", os.getcwd())

#load data
Data= pd.read_excel("PartCTelecomOne.xlsx", index_col ='Plant No')
#print (Data) #Shows if the data was taked correctly

suppliers = Data.index.tolist()[:-1] #Turn the index row in a list [:-1] delete the last 2 rows.
#print(suppliers)

markets = Data.columns.tolist()[:-2] #Turn the column names in list [:-1] delete the last coolumn
#print(markets)

unit_costs = Data.loc[suppliers, markets] #Extract the daraframe of cost unit matriz between every supplier and marke, .loc select by row and column names.
#print(unit_costs)

capacity = Data.loc[suppliers,'Capacity'] #Extract the column capacity for every supplier
#print(capacity)

fixed_cost = Data.loc[suppliers,'Fixed Cost'] #Extract the column fixed cost for every supplier
#print(fixed_cost)

demand = Data.loc["Demand", markets] #Extract demand
#print (demand)

#Initialise class
from pulp import * #importation of functions PulP
prob = LpProblem("Optimization",LpMinimize) # create the optimization model called "Optimization" looking to minimize


#Define Variables
x= {(i,j): LpVariable(f"x_{i}_{j}", lowBound=0, cat="Integer") for i in suppliers for j in markets}
#y= {(i): LpVariable(f"y_{i}", lowBound=0, cat="Binary") for i in suppliers}
#print (x)
#print(y)

# Define Objective function
prob += lpSum(unit_costs.loc[i,j] * x[(i,j)] for i in suppliers for j in markets)

#define constraints
# Constraint 1 SUm of quatitiy in x is less or equal to the capacity
for i in suppliers:
    prob+= lpSum(x[(i,j)] for j in markets)<= capacity[i]

# Constraint 2 Sum of Quatity sent by market is equal to demand
for j in markets:
    prob += lpSum(x[(i,j)] for i in suppliers) == demand[j]

prob
prob.solve()
LpStatus[prob.status]

# Displaying results
for v in prob.variables():
    print(v.name,"=",v.varValue)
print ("Variable Cost =", value(prob.objective))

total_fixed_cost =0
total_fixed_cost += sum(fixed_cost[i] for i in suppliers)
print ("Fixed Cost =", total_fixed_cost)
print("Total Cost =", total_fixed_cost + value(prob.objective))
```

```python
#second part of Assesment.
!pip install pulp
import pandas as pd
import os
#print ("Current directory is:", os.getcwd())
#os.chdir ("C:/Users/Usuario/Downloads")
#print("Current director is now:", os.getcwd())

#load data
Data= pd.read_excel("PartCTelecomOptic.xlsx", index_col ='Plant No')
#print (Data) #Shows if the data was taked correctly

suppliers = Data.index.tolist()[:-1] #Turn the index row in a list [:-1] delete the last 2 rows.
#print(suppliers)

markets = Data.columns.tolist()[:-2] #Turn the column names in list [:-1] delete the last coolumn
#print(markets)

unit_costs = Data.loc[suppliers, markets] #Extract the daraframe of cost unit matriz between every supplier and marke, .loc select by row and column names.
#print(unit_costs)

capacity = Data.loc[suppliers,'Capacity'] #Extract the column capacity for every supplier
#print(capacity)

fixed_cost = Data.loc[suppliers,'Fixed Cost'] #Extract the column fixed cost for every supplier
#print(fixed_cost)

demand = Data.loc["Demand", markets] #Extract demand
#print (demand)

#Initialise class
from pulp import * #importation of functions PulP
prob = LpProblem("Optimization",LpMinimize) # create the optimization model called "Optimization" looking to minimize


#Define Variables
x= {(i,j): LpVariable(f"x_{i}_{j}", lowBound=0, cat="Integer") for i in suppliers for j in markets}
y= {i: LpVariable(f"y_{i}", lowBound=0, cat="Binary") for i in suppliers}
#print (x)
#print(y)

# Define Objective function
prob += lpSum(y[i]* fixed_cost[i] for i in suppliers) + lpSum(unit_costs.loc[i,j] * x[(i,j)] for i in suppliers for j in markets)

#define constraints
# Constraint 1 SUm of quatitiy in x is less or equal to the capacity
for i in suppliers:
    prob+= lpSum(x[(i,j)] for j in markets)<= capacity[i]* y[i]

# Constraint 2 Sum of Quatity sent by market is equal to demand
for j in markets:
    prob += lpSum(x[(i,j)] for i in suppliers) == demand[j]


prob
prob.solve()
LpStatus[prob.status]

# Displaying results
for v in prob.variables():
    print(v.name,"=",v.varValue)
print ("Total Cost =", value(prob.objective))
```
